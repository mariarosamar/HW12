```python
# Tokenizer 
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Create a list of the words
    word_tokenized = word_tokenize(text)
    # Convert the words to lowercase
    first_clean = [word.lower() for word in word_tokenized]
    # Remove the punctuation
    remove_punctuation = [word for word in first_clean if word not in punctuation]
    # Remove the stop words
    sw = set(stopwords.words('english'))
    remove_stop = [word for word in remove_punctuation if word not in sw]
    # Lemmatize Words into root words
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in remove_stop]

    return tokens
```

```python
# Creating a new tokens column for bitcoin
bitcoin_sentiment_df["tokens"] = bitcoin_sentiment_df['text'].apply(lambda x: tokenizer(x))
```

```python
# NGrams and Frequency Analysis
```

```python
from collections import Counter
from nltk.util import ngrams
```

```python
# Generating the Bitcoin N-grams where N=2
```

```python
# All bitcoin tokens word joins in a list of list
bitcoin_token_words = bitcoin_sentiment_df['tokens'].tolist()
```

```python
# getting a big list to buil the N-grams 
big_list_words = []
for list_words in bitcoin_token_words:
    for word in list_words:
        big_list_words.append(word)
# big_list_words
remove_punctuation = [word for word in big_list_words if word not in punctuation]
# Remove the stop words
sw = set(stopwords.words('english'))
remove_stop = [word for word in remove_punctuation if word not in sw]
# Lemmatize Words into root words
lemmatizer = WordNetLemmatizer()
tokens_n_grams_bitcoin = [lemmatizer.lemmatize(word) for word in remove_stop]
```

```python
# Generating the Bitcoin N-grams where N=2
Counter(ngrams(tokens_n_grams_bitcoin, n=2)).most_common(20)
```

```python
# Generating the Ethereum N-grams where N=2
```

```python
ethereum_token_words = ethereum_sentiment_df['tokens'].tolist()
```

```python
# getting a big list to build the N-grams 
big_list_words = []
for list_words in ethereum_token_words:
    for word in list_words:
        big_list_words.append(word)
# big_list_words
remove_punctuation = [word for word in big_list_words if word not in punctuation]
# Remove the stop words
sw = set(stopwords.words('english'))
remove_stop = [word for word in remove_punctuation if word not in sw]
# Lemmatize Words into root words
lemmatizer = WordNetLemmatizer()
tokens_n_grams_ethereum = [lemmatizer.lemmatize(word) for word in remove_stop]
```

```python
# Generating the Ethereum N-grams where N=2
Counter(ngrams(tokens_n_grams_ethereum, n=2)).most_common(20)
```

```python
# Using the token_count function to generate the top 10 words from each coin
def token_count(tokens_n_grams, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens_n_grams).most_common(N)
```

```python
# Getting the top 10 words for Bitcoin
token_count(tokens_n_grams_bitcoin)   
```

```python
# Getting the top 10 words for Ethereum
token_count(tokens_n_grams_ethereum)   
```
